```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Accessing 'home_team_name' and 'away_team_name' from the dictionary and filtering by 'Egypt'

# For home matches
egypt_home_matches = df_matches['home_team'].apply(lambda x: x.get('home_team_name', '')).str.contains("Egypt", case=False, na=False)

# For away matches
egypt_away_matches = df_matches['away_team'].apply(lambda x: x.get('away_team_name', '')).str.contains("Egypt", case=False, na=False)

# Filter the dataframe based on the conditions
df_egypt_home_matches = df_matches[egypt_home_matches]
df_egypt_away_matches = df_matches[egypt_away_matches]

```

```python
df_egypt_home_matches
```

```python
df_egypt_away_matches
```

```python
# Home Matches Goals 
plt.figure(figsize = (12,4))
sns.displot(x = 'home_score', data = df_egypt_home_matches, bins = 20 )


```

```python
# Probability of scoring number of goals when playing at home
plt.figure(figsize = (12,4))
sns.kdeplot(x = 'home_score', data = df_egypt_home_matches)
```

```python
# Away Matches Goals
plt.figure(figsize = (12,4))
sns.displot(x = 'away_score', data = df_egypt_away_matches, bins = 20 )
```
